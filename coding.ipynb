```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv

#dilakukan import library, matplotlib untuk visualisasi data, numpy untuk perhitungan matematis, dan inv untuk menghitung invers matrix
```

```python
z = [5, 16, 25, 40, 50, 60, 70, 80, 90, 100]
T = [35.4, 50.1, 77.3, 92.3, 137.6, 147, 180.8, 182.7, 188.5, 223.2]

#menginput data yang akah dihitung
```

```python
G = np.zeros((len(z),2))

for i in range(10):
    G[i,0]=1
    G[i,1]=z[i]

d = np.reshape(T,(10,1))

#membuat matrix dengan 2 kolom, perulangan for menyatakan kolom pertama diisi angka 1 sedangkan kolom kedua merupakan nilai z, d mengubah data T menjadi kolom
```

```python
m = np.mat(inv(np.mat(G.transpose())*np.mat(G)))*np.mat(G.transpose())*np.mat(d)
m = m.tolist()

#mencari persamaan linier 
```

```python
a0=m[0][0]
a1=m[1][0]

linear = []

for i in range(len(z)):
    linear.append(a0 + a1*z[i])

#menghitung dan menyimpan nilai hasil perhitungan model, perulangan for menghitung nilai prediksi untuk setiap data z
```

```python
print(a0)
print(a1)

#menampilkan hasil
```

```python
plt.plot(z,T, 'r*', label='Sebaran data')
plt.plot(z, linear, label='Solusi Inversi')
plt.grid()
plt.legend()
plt.xlabel('Kedalaman(m)')
plt.ylabel('Suhu (celcius)')
plt.title('Variasi Suhu vs Kedalaman')
plt.show()

#menampilkan grafik perbandingan antara sebaran data dan solusi inversi
```

```python
h0 = 5
d = np.reshape(h-h0, (len(h),1))

G = np.zeros((len(t),2))

for i in range(len(t)):
    G[i,0]=t[i]
    G[i,1]=t[i]**2

#nilai awal  adalah 5, dan setiap nilai h data tinggi dikoreksi terhadap nilai 5, hasilnya berupa matrix kolom. g digunakan untuk membuat matrix dengan jumlah baris menyesuaikan t, dengan g merupakan kolom pertama dan g setelahnya merupakan nilai kuadrat dari t.
```

```python
GT = G.transpose()
GTG = np.matmul(GT,G)
GTG_inv = inv (GTG)
GTd = np.matmul(GT,d)

m = np.matmul(GTG_inv, GTd)

m1 = m[0][0]
m2 = m[1][0]

V0 = m1
g = -2*m2

print(m1)
print(m2)

#dilakukan trnaspose matrix, dilakukak perkalian, lalu menghitung inverse matrix, menghitung parameter model m, vo sebgaai kecepatan awal dan g sebagai percepatan gravitasi, terakrih tampilkan hasil perhitungan
```

```python
h_model = []

for i in range(len(t)):
    h_model.append(h0 +m1*t[i] + m2*(t[i]**2))

#menghitung nilai tinggi model berdasarkan persamaan glbb
```
